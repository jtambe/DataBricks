```
file_path = "dbfs:/databricks-datasets/flights/departuredelays.csv"
file_type = "csv"
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

flights_df = spark.read.format(file_type) \
.option("inferschema", infer_schema ) \
.option("header", first_row_is_header) \
.option("sep", delimiter) \
.load(file_path) \

display(flights_df)
```

```
destination_list = [row.destination for row in flights_df.select("destination").distinct().collect()]
#display(destination_list)
destination_list.sort()
print(destination_list)
```

Create Widget Dropdown

```
dbutils.widgets.dropdown(name="destination_dropdown", defaultValue="ATL", choices=destination_list, label="destinationDROPDOWN")
cur_selected_destination = dbutils.widgets.get("destination_dropdown")
print(cur_selected_destination)
```

```
dbutils.widgets.text(name="destination_text", defaultValue="ATL", label="destinationTEXT")
cur_selected_text_destination = dbutils.widgets.get("destination_text")
print(cur_selected_text_destination)
```

```
dbutils.widgets.multiselect(name="destination_multiselect", defaultValue="ATL", choices=destination_list, label="destinationMULTISELECT")
cur_multiselect_destination = dbutils.widgets.get("destination_multiselect")
print(cur_multiselect_destination)
```

```
dbutils.widgets.combobox(name="destination_combobox", defaultValue="ATL", choices=destination_list, label="destinationCOMBOBOX")
cur_combobox_destination = dbutils.widgets.get("destination_combobox")
print(cur_combobox_destination)
```

```
from pyspark.sql.functions import col
display(flights_df.filter(col("destination") == dbutils.widgets.get("destination_text")))
```
